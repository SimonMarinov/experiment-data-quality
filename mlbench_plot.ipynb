```r
if(!require('mlbench')) {
  install.packages('mlbench')
  library('mlbench')
}
```

```r
p <- mlbench.hypercube(10000,3,5,0.2)
plot(p)
library("lattice")
cloud(x.3~x.1+x.2, groups=classes, data=as.data.frame(p))
```

```r
p<-mlbench.ringnorm(1000, d=2)
plot(p)
```

```r
p<-mlbench.shapes(n=1000)
plot(p)
```

```r
p <- mlbench.simplex()
plot(p)
library("lattice")
cloud(x.3~x.1+x.2, groups=classes, data=as.data.frame(p))
```

```r
p<-mlbench.smiley()
plot(p)
```

```r
# 1 cycle each, no noise
p<-mlbench.spirals(300)
plot(p)
#
# 1.5 cycles each, with noise
#p<-mlbench.spirals(300,1.5,0.05)
#plot(p)
```

```r
p<-mlbench.threenorm(1000, d=2)
plot(p)
```

```r
p<-mlbench.twonorm(1000, d=2)
plot(p)
```

```r
p<-mlbench.waveform(100)
plot(p)
```

```r
# 2d example
p<-mlbench.xor(300,2)
plot(p)
#
# 3d example
p<-mlbench.xor(300,3)
plot(p)
```
