```python
import os
import pandas as pd
import pickle

# Path to your data directory
data_dir = 'data/train'
def_names_dir = 'data/test'

# Initialize an empty dictionary to store your data
data_dict = {}

# List all files in the directory
files = os.listdir(data_dir)
def_names_dir = os.listdir(def_names_dir)


non_poluted_dfs = []
# Iterate over the files
for file_name in def_names_dir:

    with open(os.path.join(data_dir, file_name), 'rb') as f:
        non_poluted_dfs.append((file_name, pickle.load(f)))
```

```python
comp = {}

for name,non_pol_df in non_poluted_dfs:
    name = name.split('.pkl')[0]
    filtered_list = [string for string in files if string.startswith(name) and '|' in string]
    
    pol_dfs = []
        
    for file_name in filtered_list:
        with open(os.path.join(data_dir, file_name), 'rb') as f:
            pol_dfs.append((file_name, pickle.load(f)))

    
    
    comp[name] = [(name, non_pol_df)] + pol_dfs

```

```python
for key,dfs in comp.items():
    name = dfs[0][0]

    non_pol_df = dfs[0][1]
    for pol_df in dfs[1:]:
        
        joined_df = non_pol_df.join(pol_df[1], lsuffix='_df1', rsuffix='_df2')

        diff_df = pd.DataFrame(index=non_pol_df.index)

        # Calculate the absolute difference between corresponding values
        for col in non_pol_df.columns:
             diff_df[col] = joined_df[col+'_df2'] - joined_df[col+'_df1']

        print(f"Difference between {name} and {pol_df[0]}: {diff_df}")
```
